```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def moving_average(arr, window_size):
    window = np.ones(window_size) / window_size
    return np.convolve(arr, window, mode='valid')
```

```python
window_size = 100
```

```python
train_loss = np.load("Results/train_loss.npy")
val_loss = np.load("Results/val_loss.npy")
train_loss = moving_average(train_loss, window_size)
val_loss = moving_average(val_loss, window_size)

train_loss_recon = np.load('Results/train_loss_recon.npy')
val_loss_recon = np.load('Results/val_loss_recon.npy')
train_loss_recon = moving_average(train_loss_recon, window_size)
val_loss_recon = moving_average(val_loss_recon, window_size)

train_loss_x = np.load("Results/train_loss_x.npy")
val_loss_x = np.load('Results/val_loss_x.npy')
train_loss_x = moving_average(train_loss_x, window_size)
val_loss_x = moving_average(val_loss_x, window_size)

train_loss_z = np.load("Results/train_loss_z.npy")
val_loss_z = np.load('Results/val_loss_z.npy')
train_loss_z = moving_average(train_loss_z, window_size)
val_loss_z = moving_average(val_loss_z, window_size)

train_loss_z1 = np.load("Results/train_loss_z1.npy")
val_loss_z1 = np.load('Results/val_loss_z1.npy')
train_loss_z1 = moving_average(train_loss_z1, window_size)
val_loss_z1 = moving_average(val_loss_z1, window_size)

train_loss_cons = np.load("Results/train_loss_cons.npy")
val_loss_cons = np.load('Results/val_loss_cons.npy')
train_loss_cons = moving_average(train_loss_cons, window_size)
val_loss_cons = moving_average(val_loss_cons, window_size)

train_loss_reg = np.load("Results/train_loss_reg.npy")
val_loss_reg = np.load('Results/val_loss_reg.npy')
train_loss_reg = moving_average(train_loss_reg, window_size)
val_loss_reg = moving_average(val_loss_reg, window_size)
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))

axs[0,0].plot(train_loss[20:], color = '#aa3474',label = "Training")
axs[0,0].plot(val_loss[20:], color = '#501d8a',label = "Validation")
axs[0,0].legend(fontsize=12)
axs[0,0].set_xlabel('Epochs', fontsize=12)
axs[0,0].set_ylabel('Loss', fontsize=12)
axs[0,0].set_title('Total Loss', fontsize=14, weight='bold')
#axs[0,0].text(0.1, 0.95, 'Total Loss', transform=axs[0,0].transAxes, va='top', ha='left', fontsize = 13, weight='bold')

axs[0,1].plot(train_loss_recon[20:], color = '#aa3474',label = "Training")
axs[0,1].plot(val_loss_recon[20:], color = '#501d8a',label = "Validation")
axs[0,1].legend(fontsize=12)
axs[0,1].set_xlabel('Epochs', fontsize=12)
axs[0,1].set_ylabel('Loss', fontsize=12)
axs[0,1].set_title('Reconstruction Loss', fontsize=14, weight='bold')
#axs[0,1].text(0.1, 0.95, 'Reconstruction Loss', transform=axs[0,1].transAxes, va='top', ha='left', fontsize = 13, weight='bold')

axs[0,2].plot(train_loss_x[20:], color = '#aa3474',label = "Training")
axs[0,2].plot(val_loss_x[20:], color = '#501d8a',label = "Validation")
axs[0,2].legend(fontsize=12)
axs[0,2].set_xlabel('Epochs', fontsize=12)
axs[0,2].set_ylabel('Loss', fontsize=12)
axs[0,2].set_title(r'Loss in $\.h$', fontsize=14, weight='bold')
#axs[0,2].text(0.1, 0.95, 'Loss in h ', transform=axs[0,2].transAxes, va='top', ha='left', fontsize = 13, weight='bold')

axs[1,0].plot(train_loss_z[0:], color = '#aa3474',label = "Training")
axs[1,0].plot(val_loss_z[0:], color = '#501d8a',label = "Validation")
axs[1,0].legend(fontsize=12)
axs[1,0].set_xlabel('Epochs', fontsize=12)
axs[1,0].set_ylabel('Loss', fontsize=12)
axs[1,0].set_title(r'Loss in $\.z$', fontsize=14, weight='bold')

axs[1,1].plot(train_loss_z1[20:], color = '#aa3474',label = "Training")
axs[1,1].plot(val_loss_z1[20:], color = '#501d8a',label = "Validation")
axs[1,1].legend(fontsize=12)
axs[1,1].set_xlabel('Epochs', fontsize=12)
axs[1,1].set_ylabel('Loss', fontsize=12)
axs[1,1].set_title(r'$z_i$ Loss', fontsize=14, weight='bold')

axs[1,2].plot(train_loss_cons[20:], color = '#aa3474',label = "Training")
axs[1,2].plot(val_loss_cons[20:], color = '#501d8a',label = "Validation")
axs[1,2].legend(fontsize=12)
axs[1,2].set_xlabel('Epochs', fontsize=12)
axs[1,2].set_ylabel('Loss', fontsize=12)
axs[1,2].set_title('Consistency Loss', fontsize=14, weight='bold')

plt.tight_layout()
#plt.subplots_adjust(hspace=0.1, wspace=0.3)
plt.rcParams['font.family'] = 'Times New Roman'
plt.show()
fig.savefig("Figures/Losses_all.png",dpi=300)
```
